### Quero saber se as entregas foram realizadas dentro do prazo.

```python
#Criei uma nova coluna e subtrau a data de entrega do cliete pela data estimada
merged_tres_df['entrega_no_prazo'] = merged_tres_df['data_entrega_cliente'] <= merged_tres_df['data_estimada_entrega']
print(merged_tres_df[['data_entrega_cliente', 'data_estimada_entrega', 'entrega_no_prazo']].head())
```

```python
entrega_counts = merged_tres_df['entrega_no_prazo'].value_counts()
entrega_counts
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

entrega_percent = (entrega_counts / entrega_counts.sum()) * 100

cores = ['green' if x else 'red' for x in entrega_percent.index]

plt.figure(figsize=(8, 6))
sns.barplot(x=entrega_percent.index, y=entrega_percent.values, palette=cores, order=[True, False])

for i, v in enumerate(entrega_percent):
    plt.text(i, v + 1, f'{v:.2f}%', ha='center', fontsize=12)

plt.title('Entregas dentro e fora do prazo (%)', fontsize=16)
plt.xlabel('Entrega no prazo', fontsize=12)
plt.ylabel('Porcentagem de Entregas (%)', fontsize=12)
plt.xticks([0, 1], ['Dentro do Prazo', 'Fora do Prazo'])  # Renomear os eixos

plt.show()
```

Com esse gráfico podemos perceber que a Olist possui um sistema de entregas satisfatorio com entregas no prazo de 89%.


### Quero saber qual a maior receita por tipo de pagamento

Vamos fazer essa visualização por quantidade de parcelas e por tipo de pagamento

```python
import pandas as pd

receita_por_pagamento = merged_tres_df.groupby('parcelas_pagamento')['valor_pagamento'].sum().reset_index()
receita_por_pagamento.columns = ['Parcelas de Pagamento', 'Receita Total']
print(receita_por_pagamento)
```

```python
# Quantidade de parcelas vamos executar um grafico de linhas
plt.figure(figsize=(10, 6))
lineplot = sns.lineplot(x='Parcelas de Pagamento', y='Receita Total', data=receita_por_pagamento, marker='o')

for index, value in enumerate(receita_por_pagamento['Receita Total']):
    plt.text(index, value + 0.02 * value, f'R${value:,.2f}', ha='center', fontsize=5)

plt.title('Tendência da Receita por Tipo de Pagamento', fontsize=16)
plt.xlabel('Parcelas de Pagamento', fontsize=12)
plt.ylabel('Receita Total (R$)', fontsize=12)

plt.show()
```

Aqui podemos ver que os clintes da Olist tem maior preferencia por compras com menos parcelas gerando maior receita em parcelas únicas

```python
import pandas as pd

receita_por_pagamento = merged_tres_df.groupby('tipo_pagamento')['valor_pagamento'].sum().reset_index()
receita_por_pagamento.columns = ['tipo_pagamento', 'Receita Total']
print(receita_por_pagamento)
```

```python
#para verificar a receita por tipo de pagamento vamos criar um grafico de linhas horizontais

receita_por_pagamento = receita_por_pagamento.sort_values(by='Receita Total', ascending=False)
plt.figure(figsize=(8, 4))
barplot = sns.barplot(x='Receita Total', y='tipo_pagamento', data=receita_por_pagamento, palette="Blues_d")

for index, value in enumerate(receita_por_pagamento['Receita Total']):
    plt.text(value + 0.01 * value, index, f'R${value:,.2f}', va='center', fontsize=10)

plt.title('Receita por Tipo de Pagamento', fontsize=14)
plt.xlabel('Receita Total (R$)', fontsize=12)
plt.ylabel('Tipo de Pagamento', fontsize=12)

plt.tight_layout()
plt.show()
```

Analisando os dois dados podemos ver que apesar das compras serem realizadas em baixa parcelas, o cartão de crédito ainda é a preferência na forma de pagemento, sugerindo um crédito a vista mais evidente. Sendo assim, a Olist pode investir em planos fidelidades para clientes que comprar nessa forma de pagamento ou oferecer beneficios para quem compra nessa forma de pagamento o movimentando para o boleto onde geralmente há menos taxa para o vendedor.

### Quero saber se o status do pedido varia de acordo com o valor

```python

# Análise Descritiva
descricao = merged_tres_df.groupby('status_pedido')['valor_pagamento'].describe()
print("Análise Descritiva:")
print(descricao)

# Gráfico de Caixa
plt.figure(figsize=(12, 8))
sns.boxplot(x='status_pedido', y='valor_pagamento', data=merged_tres_df)
plt.title('Distribuição do Valor do Pagamento por Status do Pedido')
plt.xlabel('Status do Pedido')
plt.ylabel('Valor do Pagamento')
plt.xticks(rotation=45)  # Rotacionar labels do eixo x para melhor visualização
plt.show()
```

Na descrição e no gráfico podemos ver que o status canceled possui a maior média de valores e um desvio de padrão alto, Isso pode indicar uma variedade nos valores de pagamento para pedidos cancelas, incluindo valores bem altos devido ao desvio padrão alto. Isso sugere que existe uma demanda de cancelamento alto e que merece atenção e analises mais profundas.
